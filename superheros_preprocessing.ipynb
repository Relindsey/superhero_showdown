```python
import numpy as np

# Function to convert height to centimeters
def convert_to_cm(height_str):
    if pd.isnull(height_str) or height_str == '-':
        return np.nan  # Return NaN for missing values
    parts = height_str.split('•')
    if len(parts) == 2:  # If height is in format "X'Y'' • Z cm"
        if 'meters' in parts[1]:
            return float(parts[1].strip().split()[0])*100
        else:
            return float(parts[1].strip().split()[0])  # Return Z (the cm part)
    else:
        return np.nan  # Return NaN if format is unexpected

# Apply the conversion function to the height column
df_fixed_lists['height_cm'] = df_fixed_lists['height'].apply(convert_to_cm)

# Drop the original 'height' column as it's no longer needed
data_cleaned = df_fixed_lists.drop(columns=['height'])

# Display the first few rows of the cleaned data to verify the transformation
data_cleaned[['name', 'height_cm']]
```

```python
# Adjust the conversion function to handle numbers with commas
def convert_to_kg_adjusted(weight_str):
    if pd.isnull(weight_str) or weight_str == '-':
        return np.nan  # Return NaN for missing values
    parts = weight_str.split('•')
    if 'tons' in weight_str:
        kg_part = parts[1].strip().split()[0].replace(',', '')  # Remove commas
        kg_part=float(kg_part)*1000
        return kg_part
    if len(parts) == 2:  # If weight is in format "X lb • Y kg"
        kg_part = parts[1].strip().split()[0].replace(',', '')  # Remove commas
        return float(kg_part)  # Convert to float and return
    else:
        return np.nan  # Return NaN if format is unexpected

# Apply the adjusted conversion function to the weight column
data_cleaned['weight_kg'] = data_cleaned['weight'].apply(convert_to_kg_adjusted)

# Display the first few rows of the cleaned data to verify the transformation
data_cleaned[['name', 'height_cm', 'weight_kg']].head(50)
```

```python
data_cleaned.loc[data_cleaned['weight'].str.contains('ton')]
```

```python
# Normalize the 'height_cm' and 'weight_kg' columns
data_cleaned['height_cm_normalized'] = (data_cleaned['height_cm'] - data_cleaned['height_cm'].min()) / (data_cleaned['height_cm'].max() - data_cleaned['height_cm'].min())
data_cleaned['weight_kg_normalized'] = (data_cleaned['weight_kg'] - data_cleaned['weight_kg'].min()) / (data_cleaned['weight_kg'].max() - data_cleaned['weight_kg'].min())

# Display the first few rows of the DataFrame to verify the normalization
data_cleaned[['name', 'height_cm', 'height_cm_normalized', 'weight_kg', 'weight_kg_normalized']].head()
```

```python
# Calculate the mode for 'height_cm' and 'weight_kg' columns
height_cm_mode = data_cleaned['height_cm'].mode()[0]
weight_kg_mode = data_cleaned['weight_kg'].mode()[0]

# Fill NaN values with the mode for both columns
data_cleaned['height_cm'].fillna(height_cm_mode, inplace=True)
data_cleaned['weight_kg'].fillna(weight_kg_mode, inplace=True)

# Since we've updated the original columns, we'll need to re-normalize them
data_cleaned['height_cm_normalized'] = (data_cleaned['height_cm'] - data_cleaned['height_cm'].min()) / (data_cleaned['height_cm'].max() - data_cleaned['height_cm'].min())
data_cleaned['weight_kg_normalized'] = (data_cleaned['weight_kg'] - data_cleaned['weight_kg'].min()) / (data_cleaned['weight_kg'].max() - data_cleaned['weight_kg'].min())

# Display the first few rows to verify the changes
data_cleaned[['name', 'height_cm', 'height_cm_normalized', 'weight_kg', 'weight_kg_normalized']].head()
```

```python
def get_words(series):
    words=[]
    if type(series)!=pd.core.series.Series:
        print('This function requieres a Series as Parameter')
    else:
        for series_item in series:
            if type(series_item)==list:
                for list_item in series_item:
                    if type(list_item==str):
                        if ',' in list_item:
                             for names in list_item.split(','):
                                words.append(names)
            elif type(list_item==str):
                        words.append(list_item)
    return words
    
```

```python
def cleaning_aliases(df):
    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        list_of_names = row['aliases']
        new_list = []
        # Iterate over each name in the list
        for name in list_of_names:
            # Count the occurrences of the name in the entire DataFrame
            count = df['aliases'].apply(lambda x: x.count(name)).sum()
            # If the name appears only once, add it to the new list
            if count == 1:
                new_list.append(name)
        # Replace the 'aliases' column with the cleaned list
        df.at[index, 'aliases'] = new_list
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download('punkt')
nltk.download('stopwords')
def literary_calculator(df):
    # Concatenate text columns
    df['superpowers_text'] = df['superpowers'].apply(lambda x: ' '.join(x))
    df['all_text'] = df['history_text'] + ' ' + df['powers_text'] + ' ' + df['superpowers_text'] + ' ' + df['occupation']
    
    # Tokenization and preprocessing
    tokens = df['all_text'].apply(lambda x: nltk.word_tokenize(x.lower()))
    tokens = tokens.apply(lambda x: [word for word in x if word.isalpha() and word not in stopwords.words('english')])
    processed_text = tokens.apply(lambda x: ' '.join(x))
    
    # TF-IDF vectorization
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(processed_text)
    
    # Get feature names (words)
    feature_names = vectorizer.get_feature_names_out()
    
    # Calculate TF-IDF weights
    for index, row in df.iterrows():
        tfidf_weights = dict(zip(feature_names, tfidf_matrix[index].toarray()[0]))
        
        # Calculate literary score
        literary_score = sum(tfidf_weights[word] for word in literary_words if word in tfidf_weights)
        df.at[index, 'literary_score'] = literary_score
    
    return df.drop(['superpowers_text', 'all_text'], axis=1)  # Drop temporary columns
```
