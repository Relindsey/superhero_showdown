```python

heroes_df=pd.read_csv('Resources/superheroes_nlp_dataset.csv')
```

```python
powers_text=""
for i in range(len(heroes_df)):
    temp=heroes_df['superpowers'][i].replace('[','').replace(']','')
    if len(temp)>0:
        powers_text=powers_text+temp+' '
powers_text
```

```python
powers2=[]
text_length=len(powers_text)
for i in range(text_length):
    power_name=""
    if powers_text[i]=="'":
        i+=1
        while powers_text[i]!="'":
            power_name=power_name+powers_text[i]
            i+=1
            if i>text_length-1: break
        i+=1
        if len(power_name)>4:
            powers2.append(power_name)

powers2.sort()
```

```python
powers_set=set(powers2)
power_list=[]
for i in range(len(powers_set)):
    power_list.append(powers_set.pop())
```

```python
power_list.sort()
power_list
```
